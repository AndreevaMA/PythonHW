```python
import math
from sympy import *
from sympy.plotting import plot
import matplotlib.pyplot as plt
```

```python
x = Symbol('x')
fx = -5 * x ** 2 + 10 * x - 150
fx
```

```python
# строим график функции с помощью библиотеку sympy, ограничиваем по х от -20 до 20 (4.)

print('График функции fx:')
p = plot(fx, (x, -20, 20))
```

```python
# строим график функции с помощью библиотки matplotlib, ограничиваем по х от -20 до 20 (4.)

import matplotlib as mpl

print('График функции fx:')

f = lambda x: -5 * x ** 2 + 10 * x - 150

x = list(range(-20, 20))
y = list(map(f, range(-20, 20)))

fig, ax = plt.subplots()
line = ax.plot(x, y)
ax.grid()

x_ticks = list(range(-20, 21, 5))
x_labels = list(map(str, list(range(-20, 21, 5))))
change = plt.xticks(ticks = x_ticks, labels = x_labels)

mpl.rcParams['lines.linewidth'] = 3
mpl.rcParams['lines.linestyle'] = '--'
mpl.rcParams['lines.color'] = 'r'

one = ax.set_xlabel('x')
two = ax.set_ylabel('f(x)')
```

```python
# находим корни уравнения (1.)

a = -5
b = 10
c = -150

discr = b ** 2 - 4 * a * c
if discr < 0:
    print('Вещественных корней нет.')
    print()

sqrt_1, sqrt_2 = solve(fx)
print(f'Корни уравнения:\n{sqrt_1}\n{sqrt_2}')
```

```python
pr = diff(fx) # производная
print(pr)
```

```python
# вычисляем вершину параболы (5.)

a = -5
b = 10
c = -150

point = []
xo = -b / (2 * a)
point.append(xo)
yo = -5 * xo**2 + 10 * xo - 150
point.append(yo)
point = tuple(point)
print(f'Вершина параболы лежит в точке с координатами {point}.')
```

```python
# интервал, на котором функция возрастает и убывает (2., 3.)

print(f'Интервал, на котором функция возрастает:')
up = solve(pr > 0)
print(up)

print()

print(f'Интервал, на котором функция убывает:')
down = solve(pr < 0)
print(down)
```

```python
# промежутки, на которых f > 0 и f < 0 (6., 7.)

pos = solve(fx > 0)
neg = solve(fx < 0)

if pos:
    print(f'Промежуток, на котором f > 0:\n{pos}')
else:
    print('Промежутков, на которых f > 0, нет.')

print()

if neg:
    print(f'Промежуток, на котором f < 0:\n{neg}')
else:
    print('Промежутков, на которых f < 0, нет.')
```
